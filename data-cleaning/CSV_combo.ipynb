```python
import os
import pandas as p

data = p.DataFrame()
# Make sure you put all the csv files into a folder named 'data' in same directory as this script
for root, dirs, files in os.walk('data/',topdown=True):
    for file in files:
        if file == "TBOP Coding_Stop and Go.xlsx":
            continue
        filename = 'data/' + file
        df = p.read_excel(filename,index_col=None,header=0)
        data = data.append(df)
display(data)  
```
